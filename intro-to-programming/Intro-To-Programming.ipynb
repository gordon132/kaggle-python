# if statements

```python
def evaluate_temp(temp):
    message = "Normal temperature."
    if temp > 38:
        message = "Fever!"
    return message
```

```python
print(evaluate_temp(37))
```

```python
print(evaluate_temp(39))
```

# if...else statements

```python
def evaluate_temp_with_else(temp):
    if temp > 38:
        message = "Fever"
    else:
        message = "Normal temparature."
    return message
```

```python
print(evaluate_temp_with_else(37))
```

# "if ... elif ... else" statements

```python
def evaluate_temp_with_elif(temp):
    if temp > 38:
        message = "Fever"
    elif temp > 35:
        message = "Normal temparature."
    else:
        message = "Low temperature."
    return message
```

```python
evaluate_temp_with_elif(36)
```

```python
print(evaluate_temp_with_elif(34))
```

# Example - Calculations

```python
def get_taxes(earnings):
    if earnings < 12000:
        tax_owed = .25 * earnings
    else:
        tax_owed = .30 * earnings
    return tax_owed
```

```python
anna_taxes = get_taxes(9000)
bob_taxes = get_taxes(15000)

print(anna_taxes)
print(bob_taxes)
```

# Module Example

```python
def add_three_or_eight(x):
    if x < 10:
        new_number = 3 + x
    else:
        new_number = 8 + x
    return new_number
    
```

```python
print(add_three_or_eight(1))
```

```python
print(add_three_or_eight(11))
```

# Example - Multiple "elif" statements

```python
# the next block of code calculates the dose of medication (in milliliters) to give to a child, based on weight (in kilograms).
def get_dose(weight):
    # Dosage is 1.25 ml for anyone under 5.2 kg
    if weight < 5.2:
        dose = 1.25
    elif weight < 7.9:
        dose = 2.5
    elif weight < 10.4:
        dose = 3.75
    elif weight < 15.9:
        dose = 5
    elif weight < 21.2:
        dose = 7.5
    # Dosage is 10 ml for anyone 21.2 kg or over
    else:
        dose = 10
    return dose
```

```python
print(get_dose(12))
```

## Intro to Lists

```python
flowers = "pink primrose,hard-leaved pocket orchid,canterbury bells,sweet pea,english marigold,tiger lily,moon orchid,bird of paradise,monkshood,globe thistle"
print(type(flowers))
print(flowers)
```

```python
# To create a list, you need to use square brackets ([, ]) and separate each item with a comma.
# Every item in the list is a Python string, so each is enclosed in quotation marks.
flowers_list = ["pink primrose", "hard-leaved pocket orchid", "canterbury bells", "sweet pea", "english marigold", "tiger lily", "moon orchid", "bird of paradise", "monkshood", "globe thistle"]
print(type(flowers_list))
print(flowers_list)
```

## Lists

```python
# length
print(len(flowers_list))
```

```python
# Indexing
print("First entry:", flowers_list[0])
print("Second entry:", flowers_list[1])
print("Last entry:", flowers_list[9])
```

```python
# Slicing
print("First three entries:", flowers_list[:3])
print("Last two entries:", flowers_list[-2:])
```

```python
# Removing items
flowers_list.remove("globe thistle")
print(flowers_list)
```

```python
# Adding items
flowers_list.append("snapdragon")
print(flowers_list)
```

```python
# Lists are not just for strings
hardcover_sales = [139, 128, 172, 139, 191, 168, 170]
```

```python
print("Length of the list:", len(hardcover_sales))
print("Entry at index 2:", hardcover_sales[2])
```

```python
# You can also get the minimum with min() and the maximum with max().
print("Maximum:", max(hardcover_sales))
print("Minimum:", min(hardcover_sales))
```

```python
# To add every item in the list, use sum().
print("Total books sold in one week:", sum(hardcover_sales))
```

```python
print("Average books sold in first five days:", sum(hardcover_sales[:5])/5)
```
