## Q5

```python
# fashinably late = arrived after at least half of the guests 
# but not the very last geuest
# total guests=7, meaning that above 4 is when is to be fashionably late
# Therefore it is indexes 4 and 5 (index positions = 0,1,2,3,4,5,6) for guests 5 and 6
```

```python
def fashionably_late(arrivals, name):
    """Given an ordered list of arrivals to the party and a name, return whether the guest with that
    name was fashionably late.
    """
    index = arrivals.index(name)
    total = len(arrivals)
    return index >= total / 2 and index != total - 1
    pass
```

```python
arrivals = ['Adela', 'Fleda', 'Owen', 'May', 'Mona', 'Gilbert', 'Ford']
result = fashionably_late(arrivals, 'Ford')
print(result)
```

```python
def has_lucky_number(nums):
    """Return whether the given list of numbers is lucky. A lucky list contains
    at least one number divisible by 7.
    """
    for num in nums:
        if num % 7 == 0:
            return True
    return False
    
ages = [1,3,4,7,9,7,6,5,2,3,7,5,6,7,9,2]
result = has_lucky_number(ages)
print(result)
```

```python
# Here's a one-line version using a list comprehension with Python's any function (you can read about what it does by calling help(any)):
def has_lucky_number(nums):
    return any([num % 7 == 0 for num in nums])

ages = [1,3,4,7,9,7,6,5,2,3,7,5,6,7,9,2]
has_lucky_number(ages)
```

```python
# Implement a function that reproduces this behaviour, returning a list of booleans corresponding to whether the corresponding element is greater than n.
def elementwise_greater_than(L, thresh):
    """Return a list with the same length as L, where the value at index i is 
    True if L[i] is greater than thresh, and False otherwise.
    
    >>> elementwise_greater_than([1, 2, 3, 4], 2)
    [False, False, True, True]
    """
    result = []
    for i in L:
        if i > thresh:
            result.append(True)
        else:
            result.append(False)
    return result

L=[1, 2, 3, 4]
elementwise_greater_than(L, 2)
```

```python
# Comprehension solution
def elementwise_greater_than(L, thresh):
    return [x > thresh for x in L]

L=[1, 2, 3, 4]
elementwise_greater_than(L, 2)
```

```python
def menu_is_boring(meals):
    """Given a list of meals served over some period of time, return True if the
    same meal has ever been served two days in a row, and False otherwise.
    """
    for meal in range(len(meals)-1):
        if meals[meal] == meals[meal+1]:
            return True
    return False

list = ['biryani', 'pilau', 'chapati', 'chapati', 'beef', 'pork', 'fish']    
menu_is_boring(list)
```

```python
# Complete the function below to help her filter her list of articles using a specific keyword
def word_search(doc_list, keyword):
    """
    Takes a list of documents (each document is a string) and a keyword. 
    Returns list of the index values into the original list for all documents 
    containing the keyword.
    """
    res = []
    for i, doc in enumerate(doc_list):
        # Split the string doc into a list of words (according to whitespace)
        words = doc.split()
        new = [word.rstrip('.,').lower() for word in words]
        # Make a transformed list where we 'normalize' each word to facilitate matching.
        # Periods and commas are removed from the end of each word, and it's set to all lowercase.
        if keyword.lower() in new:
            res.append(i)
    return res
        
```

```python
word_search(doc_list, 'casino')
```

## Question 3

```python
def multi_word_search(doc_list, keywords):
    """
    Takes list of documents (each document is a string) and a list of keywords.  
    Returns a dictionary where each key is a keyword, and the value is a list of indices
    (from doc_list) of the documents containing that keyword
    """
    res = {}
    for keyword in keywords:
        matches = []
        for i, doc in enumerate(doc_list):
            words = doc.split()
            new = [word.rstrip('.,').lower() for word in words]
            if keyword.lower() in new:
                matches.append(i)
        res[keyword] = matches
    return res
```

```python
doc_list = ["The Learn Python Challenge Casino.", "They bought a car and a casino", "Casinoville"]
keywords = ['casino', 'they']
multi_word_search(doc_list, keywords)
```

```python
# another solution
def multi_word_search(documents, keywords):
    keyword_to_indices = {}
    for keyword in keywords:
        keyword_to_indices[keyword] = word_search(documents, keyword)
    return keyword_to_indices
```

```python
def best_items(racers):
    winner_item_counts = {}
    for i in range(len(racers)):
        racer = racers[i] # racer is the dictionary at that index
        if racer['finish'] == 1: # only look at racers who fineshed 1st
            for item in racer['items']:
                if item not in winner_item_counts:
                    winner_item_counts[item] = 0
                    winner_item_counts[item] += 1

        if racer['name'] is None:
            print("WARNING: Encountered racer with unknown name on iteration {}/{} (racer = {})".format(
                i+1, len(racers), racer['name']))
    return winner_item_counts
```

```python
sample = [
    {'name': 'Peach', 'items': ['green shell', 'banana', 'green shell',], 'finish': 3},
    {'name': 'Bowser', 'items': ['green shell',], 'finish': 1},
    {'name': None, 'items': ['mushroom',], 'finish': 2},
    {'name': 'Toad', 'items': ['green shell', 'mushroom'], 'finish': 1},
]
best_items(sample)
```

```python
def blackjack_hand_greater_than(hand_1, hand_2):
    # numeric total for one hand
    # Helper function
    def hand_total(hand):
        total = 0
        aces = 0
        
        for card in hand:
            if card == "A":
                aces += 1
            elif card in ['J', 'Q', 'K']:
                total += 10
            else:
                total += int(card)

        
        for i in range(aces): # Iterates over the aces at the same time adding points
            if total + 11 <= 21: # Checks if adding 11 to the total is < 21, if less add 11
                total += 11
            else:                # if more add instead 1
                total += 1
        return total

    total_1 = hand_total(hand_1)
    total_2 = hand_total(hand_2)

    if total_1 > 21: # First rule
        return False
        
    return total_1 > total_2 or total_2 > 21 # Second rule
```
